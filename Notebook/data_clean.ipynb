```python
# Which department has the highest attrition?
department_attrition = df.groupby('Department')['Attrition'].value_counts().reset_index()
department_attrition
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Department', y='count', hue='Attrition', data=department_attrition)
plt.title('Attrition by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
# Which job role has the highest salary?
role_salary= df.groupby('JobRole')['MonthlyIncome'].max().reset_index()
plt.figure(figsize=(20,6))
sns.barplot(x='JobRole', y='MonthlyIncome', data=role_salary)
plt.title('Monthly Income by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Monthly Income')
plt.show()
```

```python
# Which department has the most employees?
dempartment_employees = df.groupby('Department')['EmployeeCount'].count().reset_index()
plt.figure(figsize=(15,6))
sns.barplot(x='Department', y='EmployeeCount', data=dempartment_employees)
plt.title('Employee Count by Department')
plt.xlabel('Department')
plt.ylabel('Employee Count')
plt.show()
```

```python
# Male vs Female employees
gender_employees = df.groupby('Gender')['EmployeeCount'].count().reset_index()
plt.figure(figsize=(15,6))
plt.pie(gender_employees['EmployeeCount'], labels=gender_employees['Gender'], autopct='%1.1f%%')
plt.title('Employee Count by Gender')
plt.show()
```

```python
# Avgerage Monthly Income by Marital Status
marital_income = df.groupby('MaritalStatus')['MonthlyIncome'].mean().reset_index()
plt.figure(figsize=(15,6))
sns.barplot(x='MaritalStatus', y='MonthlyIncome', data=marital_income)
plt.title('Average Monthly Income by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Average Monthly Income')
plt.show()
```
